## Another approach to finding links

Each time we look for something that 'Looks like ___', 
be it address, link or phone number 
we can use regular expressions to find them.

Regular expressions are super usefull when needed, 
but that happens very rarely. 
As they are complicated to learn and easy to forget it's best to learn them on the fly when needed.

```python
import re
```

```python
# Find all occurances of `<a href="/locations/london/balham`
re.findall(r'<a href="/locations/london/balham">', request.text)
```

```python
# Find all occurance of "Looks like a link"
re.findall(r'<a href="/locations/london/[a-z-]*">', request.text)
```

```python
# Remove duplicates
urls=list(set(re.findall(r'<a href="/locations/london/[a-z-]*">', request.text)))
urls[0]
```

```python
url = urls[0]
url.split('/')[-1].split('"')[0]
```

```python
neighborhood_names = [url.split('/')[-1].split('"')[0] for url in urls]
neighborhood_names
```

Mission accomplished! :)

## Finding all tags for a given neighborhood

```python
neighborhood_names[0]
```

```python
url = 'https://www.airbnb.com/locations/london/{}'.format(neighborhood_names[0])
url
```

```python
request = requests.get(url)
webpage_text = request.text
webpage_text
```

```python
results = webpage_text.split('<div class="neighborhood-tag">')
results = results[1:]
results[0]
```

```python
result = results[0]
result.split('<')[0].strip()
```

```python
tags = [result.split('<')[0].strip() for result in results]
tags
```

```python
tags_dict = {}
for name in neighborhood_names:
    url = 'https://www.airbnb.com/locations/london/{}'.format(neighborhood_names[0])
    request = requests.get(url)
    webpage_text = request.text
    results = webpage_text.split('<div class="neighborhood-tag">')[1:]
    tags = [result.split('<')[0].strip() for result in results]
    tags_dict[name] = tags
```
